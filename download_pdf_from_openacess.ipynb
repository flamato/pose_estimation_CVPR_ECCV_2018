```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
# Created by wjbKimberly on 18-05-23

import re
import urllib.request
import os
import json
import random
import time
import requests


def download(url, save_path):
    contents=requests.get(url,stream=True)
    f = open(save_path, 'wb')
    for chunk in contents.iter_content(chunk_size=512):
        if chunk:
            f.write(chunk)
    f.close()

def get_url_list(web_url,conference,keywords):
    patt_prefix=r'<a href="(content_.+?/papers/.+?.pdf)">pdf</a>'
    patt_name=r'<a href="content_%s_2018/html/.+?\.html">(.+?)</a>'%conference
    html_sample = str(urllib.request.urlopen(web_url).read().decode('utf-8'))
    paper_prefix_list=re.findall(patt_prefix, html_sample)
    paper_name_list=re.findall(patt_name, html_sample)
    other_url_list=[]
    pose_url_list=[]
    for prei,ni in zip(paper_prefix_list, paper_name_list):
        ni=ni.replace(":","：")
        ni=ni.replace("/"," or ")
        ni=ni.replace("?","？")
        for ki in keywords:
            if ki in ni:
                pose_url_list.append((prei,ni))
                break
        else:
            other_url_list.append((prei,ni))
    return other_url_list,pose_url_list

if __name__ == '__main__':
    conference="ECCV"
    keywords=["Pose","pose"]
    subset_name="pose"
    
    root_path="http://openaccess.thecvf.com/%s2018.py"%conference
    other_url_list,target_url_list=get_url_list(root_path,conference,keywords)
    print(len(other_url_list),len(target_url_list))
    # download papers from other domains
    for i,pi in enumerate(target_url_list):
        urli="http://openaccess.thecvf.com/"+pi[0]
        download(urli,"%s/"%subset_name+pi[1]+".pdf")
        print(i,pi[1])
    # download papers from target domains
    for i,oi in enumerate(other_url_list):
        urli="http://openaccess.thecvf.com/"+oi[0]
        download(urli,oi[1]+".pdf")
        print(i,oi[1])
```
